```python
df.columns = df.columns.str.replace("_x0020_","_").str.strip().str.lower()
```

```python
df.columns
```

```python
df["arrival_date"] = pd.to_datetime(df["arrival_date"],errors = "coerce")
```

```python
df.info()
```

```python
df["arrival_date"].isnull().sum()
```

```python
price_col = ["min_price", "max_price", "modal_price"]
for col in price_col:
    df[col] = pd.to_numeric(df[col],errors = "coerce")
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df = df[df["min_price"] <= df["modal_price"]]
df = df[df["modal_price"] <= df["max_price"]]
```

```python
df["price_range"] = df["max_price"] - df["min_price"]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["avg_price"] = (df["max_price"] - df["min_price"]) / 2
```

```python
df.columns
```

```python
df.sample(20)
```

```python
state_price_summary = df.groupby("state").agg(avg_modal_price = ("modal_price","mean"),
                                              max_modal_price = ("modal_price","max"),
                                              min_modal_price = ("modal_price","min")).reset_index()
state_price_summary
```

```python
validity_threshold = df["price_range"].median()

df["validity_level"] = np.where(df["price_range"] > validity_threshold, "High","Low")
df
```

```python
df["market_price_rank"] = df.groupby(["commodity","state"])["modal_price"].rank(ascending = False)
df
```
